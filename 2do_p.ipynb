```python
import torch
import cv2
import numpy as np
```

```python
modelo = torch.hub.load('ultralytics/yolov5', 'custom',
                       path = 'C:/Users/Loren/Desktop/Facu/Inteligencia Artificial/visual_IA/2do_parcial/2do_parcial_IA/modelo_poker.pt')
```

# camara

```python
# Realizo Videocaptura
cap = cv2.VideoCapture(1)

# Empezamos
t=0
while cap.isOpened():
    ret, frame = cap.read()
    detect = modelo(frame)

    info = detect.pandas().xyxy[0]  
    print(info)
    cv2.imshow('Detector cartas', np.squeeze(detect.render()))

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break


cap.release()
cv2.destroyAllWindows()
```

# video

```python

capture = cv2.VideoCapture('UNO1.mkv')
salida = cv2.VideoWriter('videoSalida.avi',cv2.VideoWriter_fourcc(*'XVID'),20.0,(640,480))
while (capture.isOpened()):
    ret, frame = capture.read()
    if (ret == True):
        detect = modelo(frame)
        info = detect.pandas().xyxy[0]
        print(info)
        #cv2.imshow("uno", np.squeeze(detect.render()))
        salida.write(np.squeeze(detect.render()))
        if (cv2.waitKey(30) == ord('s')):
            break
    else:
        break

capture.release()
salida.relase()
cv2.destroyAllWindows()
```
