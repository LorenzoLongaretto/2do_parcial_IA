```python
import torch
import cv2
import numpy as np
```

```python
modelo = torch.hub.load('ultralytics/yolov5', 'custom',
                       path = 'C:/Users/Loren/Desktop/Facu/Inteligencia Artificial/visual_IA/2do_parcial/best.pt')
```

# camara

```python
# Realizo Videocaptura
cap = cv2.VideoCapture(1)

# Empezamos
t=0
while t!=27:
    # Realizamos lectura de frames
    ret, frame = cap.read()

    # Correccion de color
    #frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Realizamos las detecciones
    detect = modelo(frame)

    info = detect.pandas().xyxy[0]  # im1 predictions
    print(info)

    # Mostramos FPS
    cv2.imshow('Detector cartas', np.squeeze(detect.render()))

    # Leemos el teclado
    t = cv2.waitKey(5)


cap.release()
cv2.destroyAllWindows()
```

# video

```python
capture = cv2.VideoCapture('UNO_cortado.mp4')

while (capture.isOpened()):
    ret, frame = capture.read()
    if (ret == True):
        detect = modelo(frame)
        info = detect.pandas().xyxy[0]
        print(info)
        cv2.imshow("uno", np.squeeze(detect.render()))
        if (cv2.waitKey(30) == ord('s')):
            break
    else:
        break

capture.release()
cv2.destroyAllWindows()
```
